```python
# Data Pre-Processing - Job listing Dataset 
# Import necessary packages
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
# Reading the csv file as input
jobs = pd.read_csv("./data/Indeed_Jobs_listing.csv") 
```

```python
# Data Cleaning
# Processing hourly wage
jobs['salary'] = jobs.loc[:,'salary_formatted'].str.extract('([\d,]+)')

# Remove the comma values from the salary 
jobs['salary'] = jobs['salary'].str.replace(',','')

# Replace all the NaN to zero values
jobs['salary'] = jobs['salary'].replace(np.nan, 0)

# Convert the salary to hourly wage
def hourly_wage(x, salary, h):
    if(len(str(x[salary])) >= 5):
        x[h] = int(x[salary])/(40*52)
    elif(len(str(x[salary])) == 4):
         x[h] = int(x[salary])/(40*4)
    else:
         x[h] = int(x[salary])
    return x

jobs = jobs.apply(lambda x: hourly_wage(x, 'salary', 'hourly_wage'), axis=1)
```

```python
# Data Pre-Processing - resumes_indeed_com-job_sample dataset
resume = pd.read_csv("./data/resumes_indeed_com-job_sample.csv")
```

```python
# Locate the country corresponding to the given state/city
resume['place'] = resume.loc[:,'Location'].str.extract('([\w\s]+)')

geolocator = Nominatim(user_agent = "geoapi")

def country_search(x, p, c):
    place = geolocator.geocode(p)
    country = (place.raw['display_name']).split(',')
    x[c] = country[3].strip()
    return x

resume = resume.apply(lambda x: country_search(x, 'place', 'country'), axis=1)
```
